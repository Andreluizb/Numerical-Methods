# One-Dimensional Gaussian Filter

1-D Gaussian filter function:
$$
F(x) = \left(\frac{6}{\pi \Delta^2}\right) ^ {1/2} exp\left({-\frac{6x^2}{\Delta^2}}\right)
$$
Filtered data $\bar{y}$ can be obtained from original $y$:
$$
\bar{y}(x) = \int^{+\infty}_{-\infty} y(x') F(x-x') dx'
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Raw data
x = np.linspace(-1., 1., 101)
y = np.ones_like(x)
y[np.where(x >= 0.0)] = 2.0
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Gaussian weights
def gauss_weight(x, a, delta):
    gw =  math.sqrt(6./math.pi/delta**2) * \
          math.exp(-6.*(x-a)**2/delta**2)
    return gw
```

```python
gw = np.zeros_like(x)
x0 = 0.0
delta = 0.5
for i in range(len(x)):
    gw[i] = gauss_weight(x[i], x0, delta)
```

```python
plt.plot(x, gw)
plt.xlabel('x')
plt.ylabel('Gaussian Func')
plt.show()
```

We can see that $\Delta$ is half of the curve width

**The integral of the entire gauss function is unity.**

```python
s = 0.
for i in range(len(x)-1):
    s = s + gw[i] * (x[i+1] - x[i])
print('Integral of gaussian function:', s)
```

However, it is not the case if the centre of the curve is shifted.

```python
x0 = 0.8
for i in range(len(x)):
    gw[i] = gauss_weight(x[i], x0, delta)
s = 0.
for i in range(len(x)-1):
    s = s + gw[i] * (x[i+1] - x[i])
print('Integral of gaussian function:', s)
```

```python
plt.plot(x, gw)
plt.xlabel('x')
plt.ylabel('Gaussian Func')
plt.show()
```

```python
# Integrate
def gaussian_filter(x, y, delta):
    yf = np.zeros_like(y)
    for i in range(len(x)-1):
        gw = 0.
        for j in range(len(x)-1):
            gw1 = gauss_weight(x[j], x[i], delta)
            gw2 = gauss_weight(x[j+1], x[i], delta)
            yf[i] = yf[i] + 0.5*(gw1*y[j] + gw2*y[j+1]) * (x[j+1] - x[j])
            gw = gw + 0.5 * (gw1 + gw2) * (x[j+1] - x[j])
    return yf
```

```python
delta = 0.5
yf = np.zeros_like(y)
yf = gaussian_filter(x, y, delta)
plt.plot(x, y, 'b--', label='Original')
plt.plot(x, yf, 'r-', label='Filtered')
plt.xlabel('x')
plt.ylabel('y or yf')
plt.legend(loc='best')
plt.show()
```

What happened? The filtered curve seems okay in the middle but is off when it is close to the two ends.

**Recall that the gaussian function is not complete (integral is not unity) when approaching the boundaries.**

One solution is to take into account the weights close to boundary.

Another solution is to extend the domain in both directions by at least $\Delta$

```python
# Integrate
def gaussian_filter2(x, y, delta):
    yf = np.zeros_like(y)
    for i in range(len(x)-1):
        gw = 0.
        for j in range(len(x)-1):
            gw1 = gauss_weight(x[j], x[i], delta)
            gw2 = gauss_weight(x[j+1], x[i], delta)
            yf[i] = yf[i] + 0.5*(gw1*y[j] + gw2*y[j+1]) * (x[j+1] - x[j])
            gw = gw + 0.5 * (gw1 + gw2) * (x[j+1] - x[j])
        yf[i] = yf[i] / gw
    return yf
```

```python
delta = 0.5
yf = np.zeros_like(y)
yf = gaussian_filter2(x, y, delta)
plt.plot(x, y, 'b--', label='Original')
plt.plot(x, yf, 'r-', label='Filtered')
plt.xlabel('x')
plt.ylabel('y or yf')
plt.legend(loc='best')
plt.show()
```

This is much better. But the right boundary is wrong because integral is not performed at it. Fix this by extrapolation.

```python
# Integrate
def gaussian_filter3(x, y, delta):
    yf = np.zeros_like(y)
    for i in range(len(x)-1):
        gw = 0.
        for j in range(len(x)-1):
            gw1 = gauss_weight(x[j], x[i], delta)
            gw2 = gauss_weight(x[j+1], x[i], delta)
            yf[i] = yf[i] + 0.5*(gw1*y[j] + gw2*y[j+1]) * (x[j+1] - x[j])
            gw = gw + 0.5 * (gw1 + gw2) * (x[j+1] - x[j])
        yf[i] = yf[i] / gw
        yf[-1] = yf[-2]
    return yf
```

```python
delta = 0.5
yf = np.zeros_like(y)
yf = gaussian_filter3(x, y, delta)
plt.plot(x, y, 'b--', label='Original')
plt.plot(x, yf, 'r-', label='Filtered')
plt.xlabel('x')
plt.ylabel('y or yf')
plt.legend(loc='best')
plt.show()
```

This is what we want.

Note that the actual filter operation changed to:
$$
\bar{y}(x) = 
\frac{\int^{+\infty}_{-\infty} y(x') F(x-x') dx'}
{\int^{+\infty}_{-\infty} F(x-x') dx'}
$$

```python
# Integrate
def gaussian_filter4(x, y, delta):
    # Grid size
    dx = np.zeros(len(x)-1)
    dx = x[1:] - x[:-1]
    # Use the smallest dx for the extended mesh. In this example, the mesh is uniform
    n_ex = int(delta / min(dx)) # number of points in the extended region
    # New x
    x_ex = np.zeros(len(x)+2*n_ex)
    x_ex[n_ex:len(x)+n_ex] = x
    for i in range(n_ex):
        x_ex[i] = x[0] - (n_ex-i)*min(dx)
    for i in range(len(x)+n_ex, len(x_ex)):
        x_ex[i] = x[-1] + (i-len(x)-n_ex+1)*min(dx)
    # Integrate
    ynew = np.zeros_like(x_ex)
    ynew[n_ex:len(x)+n_ex] = y
    for i in range(n_ex, len(x+n_ex)):
        for j in range(len(x_ex)-1):
            if j > n_ex and j < len(x)+n_ex:
                gw1 = gauss_weight(x_ex[j], x_ex[i], delta)
                gw2 = gauss_weight(x_ex[j+1], x_ex[i], delta)
                ynew[i] = ynew[i] + 0.5*(gw1*ynew[j] + gw2*ynew[j+1]) * (x_ex[j+1] - x_ex[j])
    yf = np.zeros_like(y)
    yf = ynew[n_ex:len(x)+n_ex]
    return yf
```
